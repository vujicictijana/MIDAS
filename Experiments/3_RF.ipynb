```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import time
import bz2
import _pickle as cPickle
from sklearn.metrics import accuracy_score
import numpy as np


my_os = "mac" #win mac

separator = "\\"

if my_os == "mac":
    separator = "//"

mode = "B"
# empty, B, C

type = "AD"
# AD, AC

file_name = "algorithms" + separator + "LR_" + type

if mode == "A":
    df_train = pd.read_pickle("train1.pkl")
    df_train = pd.concat([df_train, pd.read_pickle("train2.pkl")])
    df_train = pd.concat([df_train, pd.read_pickle("train3.pkl")])
    df_train = pd.concat([df_train, pd.read_pickle("train4.pkl")])
    df_train = pd.concat([df_train, pd.read_pickle("train5.pkl")])
    df_val = pd.read_pickle("val.pkl")
if mode == "B" or mode == "C":
    file_name = file_name + "_" + mode
    df_train = pd.read_pickle("data" + separator + "train1_" + mode + ".pkl")
    df_train = pd.concat([df_train, pd.read_pickle("data" + separator + "train2_" + mode + ".pkl")])
    df_train = pd.concat([df_train, pd.read_pickle("data" + separator + "train3_" + mode + ".pkl")])
    df_train = pd.concat([df_train, pd.read_pickle("data" + separator + "train4_" + mode + ".pkl")])
    df_train = pd.concat([df_train, pd.read_pickle("data" + separator + "train5_" + mode + ".pkl")])
    df_val = pd.read_pickle("val_B.pkl")

y_train = df_train['Anomaly']
X_train = df_train.drop(['Anomaly'],axis= 1)

y_train=y_train.astype('int')

y_val = df_val['Anomaly']
y_val=y_val.astype('int')

X_val = df_val.drop(['Anomaly'],axis= 1)
X_val = X_val.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)

if type == "AD":
    y_train = y_train.replace(2,1);
    y_train  = y_train.replace(3,1);
    y_val = y_val.replace(2,1);
    y_val  = y_val.replace(3,1);

y_val = y_val.array
y_train = y_train.array


x = range(5, 50, 5)
max = 0
maxAcc = 0
for n in x:
    rf = RandomForestClassifier(n_estimators = n)
    rf.fit(X_train, y_train)
    prediction = rf.predict(X_val)
    accuracy = accuracy_score(prediction, y_val)
    print("No. of trees: ", n)
    print("Accuracy: ", accuracy)
    if maxAcc<accuracy:
        max = n
        maxAcc = accuracy

result = "RF: \n"
result += "Best no. of trees: " + str(max) + "\n"
result += "Best accuracy: " + str(maxAcc) + "\n"
print(result)

rf = RandomForestClassifier(n_estimators = max)
start_time = time.time()
rf.fit(X_train, y_train)
my_time = (time.time() - start_time)

f = bz2.BZ2File(file_name + ".pbz2", "wb")
cPickle.dump(rf,f)
f.close()

result += "Time train (s): " + str(my_time)

f1 = open(file_name + ".txt", "w")
f1.write(result)
f1.close()
```
