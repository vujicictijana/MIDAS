```python
import pandas as pd
import os
import time
import numpy as np
from sklearn.preprocessing import MinMaxScaler


current = os.getcwd()
path = os.path.dirname(current) + '/Dataset/';


my_os = "mac" #win mac

separator = "\\"

if my_os == "mac":
    separator = "//"

files = os.listdir(path)
n = len(files)

df_temp = pd.read_csv(path + files[0],sep = ',', dtype=str)
df_temp = df_temp[0:0]
start = time.time()

type = "train" # train,val,test

# 1,2,3,4,5
no = 1

border_for_validation1 = 501
border_for_validation2 = 601

if no == 1:
    border_for_train1 = 0
    border_for_train2 = 101
    border_for_testing1 = 601
    border_for_testing2 = 701
if no == 2:
    border_for_train1 = 101
    border_for_train2 = 201
    border_for_testing1 = 701
    border_for_testing2 = 801
if no == 3:
    border_for_train1 = 201
    border_for_train2 = 301
    border_for_testing1 = 801
    border_for_testing2 = 901
if no == 4:
    border_for_train1 = 301
    border_for_train2 = 401
    border_for_testing1 = 901
    border_for_testing2 = 1001
if no == 5:
    border_for_train1 = 401
    border_for_train2 = 501

for i in range(0,n):
    if files[i].endswith('.csv'):
        print(i)
        run_Id =  int(files[i].split("_")[1])
        if type == "train":
            if run_Id >= border_for_train1 and run_Id<border_for_train2:
                df = pd.read_csv(path + files[i],  sep = ',', dtype=str)
                df_temp = pd.concat([df_temp, df])
        if type == "val":
            if run_Id >= border_for_validation1 and run_Id < border_for_validation2:
                df = pd.read_csv(path + files[i],  sep = ',', dtype=str)
                df_temp = pd.concat([df_temp, df])
        if type == "test":
            if run_Id >= border_for_testing1 and run_Id < border_for_testing2:
                df = pd.read_csv(path + files[i],  sep = ',', dtype=str)
                df_temp = pd.concat([df_temp, df])


end = time.time()
final_time = end - start

print("Total time to load data: "  + str(final_time))

print("Runs: " + str(len(df_temp['Run id'].unique())))
print(np.sort(df_temp['Run id'].unique()))

start = time.time()

df_temp['Anomaly'] = np.where(df_temp['Anomaly'] == '-', 0, df_temp['Anomaly'])
df_temp['Anomaly'] = np.where(df_temp['Anomaly'] == 'Freeze', 1, df_temp['Anomaly'])
df_temp['Anomaly'] = np.where(df_temp['Anomaly'] == 'Ramp', 2, df_temp['Anomaly'])
df_temp['Anomaly'] = np.where(df_temp['Anomaly'] == 'Step', 3, df_temp['Anomaly'])
Y = df_temp['Anomaly']

# print(Y.unique())

df_temp = df_temp.drop(['number','Timestamp', 'Parameter for Anomaly','Anomaly','Actual value', 'Run id'],axis= 1)
df_temp = df_temp.astype(float)
df_temp.fillna(0)
scaler = MinMaxScaler()
df_temp = pd.DataFrame(scaler.fit_transform(df_temp.values), columns=df_temp.columns, index=df_temp.index)
df_temp['Anomaly'] = Y

end = time.time()
final_time = end - start

print("Total time to preprocess data: "  + str(final_time))

start = time.time()
if type == "val":
    df_temp.to_pickle("data" + separator + type + ".pkl")
else:
    df_temp.to_pickle("data" + separator + type + str(no) + ".pkl")

end = time.time()
final_time = end - start

print("Total time write to file: "  + str(final_time))

```
